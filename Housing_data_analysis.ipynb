```python
tr = ColumnTransformer(transformers = [
    ('t1',OneHotEncoder(drop='first',sparse=False),['mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea']),
    ('t2',OrdinalEncoder(categories=[['furnished','semi-furnished','unfurnished']]),['furnishingstatus'])
],remainder='passthrough')
```

```python
tr
```

```python
a = tr.fit_transform(X_train)
a
a.shape
```
