```python
Q1=df['price'].quantile(0.25)
Q1
```

```python
Q3=df['price'].quantile(0.75)
Q3
```

```python
iqr=Q3-Q1
iqr
```

```python
ul=Q3+1.5*iqr
ul
```

```python
ll=Q1-1.5*iqr
ll
```

```python
df[(df['price']>ul) | (df['price']<ll)]
```

```python
df['price']=np.where((df['price']>ul),
                                  ul,
                                  np.where((df['price']<ll),
                                                        ll,
                                                        df['price']))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
transform=ColumnTransformer(transformers=[
    ('t1',MinMaxScaler(),['price']),
    ('t2',StandardScaler(),['area']),
    ('t3',OneHotEncoder(sparse=False,drop='first'),['mainroad','guestroom','basement','airconditioning']),
    ('t4',OrdinalEncoder(categories=[['unfurnished','semi-furnished','furnished']]),['furnishingstatus'])
],remainder='passthrough')
```

```python
new_df=transform.fit_transform(df)
new_df.shape
```

```python
new_df=pd.DataFrame(new_df,columns=['price','area','mainroad','guestroom','basement','airconditioning','furnishingstatus','bedrooms','bathrooms','stories','parking'])
```
