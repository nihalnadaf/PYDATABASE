###### 3.Find the district that gets the highest annual rainfall.


```python
df[['DISTRICT','ANNUAL']].sort_values(by='ANNUAL',ascending=False).head(1)
```

```python
df[['DISTRICT','ANNUAL']].sort_values(by='ANNUAL',ascending=False).head(5)
```

###### 5.Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.

```python
df.drop(['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'] , axis=1, inplace=True)
```

###### 6.Display the state-wise mean rainfall for all the months using a pivot table.


```python
st = pd.pivot_table( df, index='STATE_UT_NAME' , values=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'])
st.sample(2)
```
